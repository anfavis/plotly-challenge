```python
#Dependencies
import os

import pandas as pd
import numpy as np

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
```

```python
# Check db
db_path = "belly_button_biodiversity.sqlite"
engine = create_engine(f"sqlite:///{db_path}")
conn = engine.connect()
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# Print all of the classes/tables mapped to the Base
Base.classes.keys()
```

```python
# Assign to variables
otu_df = pd.read_sql("SELECT * FROM otu", conn)
samples_df = pd.read_sql("SELECT * FROM samples", conn)
samples_metadata_df = pd.read_sql("SELECT * FROM samples_metadata", conn)
```
